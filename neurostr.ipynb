AND GATE

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

```python
def AND(a, b):
  input = np.array([a, b])
  weights = np.array([0.4, 0.4])
  bias = -0.6
  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0

  else:return 1
```

```python
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

OR GATE

```python
def OR(a, b):
  input = np.array([a, b])
  weights = np.array([0.4, 0.5])
  bias = -0.3
  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0

  else:return 1
```

```python
print(OR(0, 0))
print(OR(0, 1))
print(OR(1, 0))
print(OR(1, 1))
```

NAND GATE

```python
def NAND(a, b):
  input = np.array([a, b])
  weights = np.array([-0.6, -0.5])
  bias = 0.7
  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0

  else:
    return 1
```

```python
print(NAND(0, 0))
print(NAND(0, 1))
print(NAND(1, 0))
print(NAND(1, 1))
```

XOR GATE

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

계단함수

```python
def step_function(x):
  if x > 0:
    return 1
  else:
    return 0
```

```python
def step_function_for_numpy(x):
  y = x > 0
  return y.astype(np.int)
```

```python
print(step_function(-3))
print(step_function(5))

a = np.array([5, 3, 4, 2.0])
print(step_function_for_numpy(a))
```

시그모이드 함수

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
print(sigmoid(3))
print(sigmoid(-3))
```

ReLU 함수

```python
def ReLU(x):
  if x > 0:
    return x
  else:
    return 0
```

```python
print(ReLU(5))
print(ReLU(-3))
```

하이퍼볼릭탄젠트 함수

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
print(tanh(3))
print(tanh(-3))
```

Softmax 함수

```python
def softmax(a):
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a
  return y
```

```python
a = np.array([0.3, 0.2, 4.0, -1.2])
print(softmax(a))
print(np.sum(softmax(a)))
```
